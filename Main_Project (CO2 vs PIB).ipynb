```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Merge PIB vs CO2 

```python
#Create C02 DF:
csv_co2 = './CO2/Resources/CO2_country.csv'
co2_df = pd.read_csv(csv_co2)

#Create PIB DF:
csv_pib = './PIB/Resources/PIB.csv'
pib_df = pd.read_csv(csv_pib)

#Merge CO2 & PIB Dataframes by two Colmuns (Country & Year):
co2_pib = pd.merge(co2_df,pib_df,how='left', left_on=['Country','Year'],right_on=['Country','Year'])

co2_pib.head()
```

Let's Plot! (CO2 vs PIB)

```python
country = input ("Please type the country you would like to analyze: ").lower()

country_plt = co2_pib[co2_pib['Country'] == country]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

# Plot the CO2 Emissions
co2, = ax1.plot(country_plt['Year'], country_plt['Total'], color="blue", label="C02 Emissions")

# Plot the PIB Growth per country
pib, = ax2.plot(country_plt['Year'], country_plt['PIB'], color="red",label='PIB')

plt.title("CO2 Emissions & PIB Growth")
ax1.set_ylabel('C02', color='b')
ax2.set_ylabel('PIB Growth (%)', color='r')
plt.grid(True)

plt.legend(handles=[co2, pib], loc="lower right")

fig.savefig('./Images/CO2vsPIB' + country + '.png')

plt.show()
```
