```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Merge PIB vs CO2 

```python
#Create C02 DF:
csv_co2 = './CO2/Resources/CO2_country.csv'
co2_df = pd.read_csv(csv_co2)

#Create PIB DF:
csv_temp = './TEMP/Resources/merged2.csv'
temp_df = pd.read_csv(csv_temp)

#Merge CO2 & TEMP Dataframes by two Colmuns (Country & Year):
co2_temp = pd.merge(co2_df,temp_df,how='left', left_on=['Country','Year'],right_on=['Country','Year'])

co2_temp.head()
```

Let's Plot! (CO2 vs PIB)

```python
country = input ("Please type the country you would like to analyze: ").lower()

country_plt = co2_temp[co2_temp['Country'] == country]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

# Plot the CO2 Emissions
co2, = ax1.plot(country_plt['Year'], country_plt['Total'], color="blue", label="C02 Emissions")

# Plot the PIB Growth per country
temp, = ax2.plot(country_plt['Year'], country_plt['AverageTemperature'], color="red",label='Temp')

plt.title("CO2 Emissions & Temperature Growth")
ax1.set_ylabel('C02', color='b')
ax2.set_ylabel('Average Temperature (C)', color='r')
plt.grid(True)

fig.savefig('./Images/CO2vsAvgTemp' + country + '.png')

plt.legend(handles=[co2, temp], loc="lower right")
plt.show()
```

```python
country = input ("Please type the country you would like to analyze: ").lower()

country_plt = co2_temp[co2_temp['Country'] == country]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

# Plot the CO2 Emissions
co2, = ax1.plot(country_plt['Year'], country_plt['Total'], color="blue", label="C02 Emissions")

# Plot the PIB Growth per country
temp, = ax2.plot(country_plt['Year'], country_plt['Standard Deviation'], color="red",label='Std Variation')

plt.title("CO2 Emissions & Climate Variation")
ax1.set_ylabel('C02', color='b')
ax2.set_ylabel('Climate Variation', color='r')
plt.grid(True)

plt.legend(handles=[co2, temp], loc="lower right")
plt.show()
```
